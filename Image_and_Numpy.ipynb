## Images and Numpy

```python
# Conjustion called PIL (Python Imaging Library)
```

```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
pic = Image.open('../Images/Photo1.jpg')
```

```python
type(pic)
```

```python
pic_arr = np.asarray(pic)
type(pic_arr)
```

```python
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
pic2 = pic_arr.copy()
```

```python
# R - Channel, G - Channel, B - Channel
plt.imshow(pic2[:, :, 0])
```

```python
# show the image pixel between 0 and 255
pic2[:, :, 0]
```

```python
# Red channel values 0 - No Red pure black -255 full pure red 
plt.imshow(pic2[:, :, 0], cmap='gray') # closer, more red
```

```python
# Green channel values 0-255
plt.imshow(pic2[:, :, 1], cmap='gray')
```

```python
# Blue channel values 0-255
plt.imshow(pic2[:, :, 2], cmap='gray')
```

```python
# Zero out Green channel
pic2[:, :, 1] = 0
plt.imshow(pic2)
```

```python
# Zero out Blue channel
pic2[:, :, 2] = 0
plt.imshow(pic2)
```
