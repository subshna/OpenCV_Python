### Harris Corner Detection

### Harris Corner Detection

```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('../Images/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_flat_chess)
```

```python
plt.imshow(gray_flat_chess, cmap='gray')
```

```python
real_chess = cv2.imread('../Images/real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_chess)
```

```python
gray_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_real_chess, cmap='gray')
```

```python
gray = np.float32(gray_flat_chess)
```

```python
dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)
```

```python
dst = cv2.dilate(src=dst, kernel=None)
```

```python
flat_chess[dst>0.01 * dst.max()] = [255, 0, 0] # RGB
```

```python
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real_chess)

dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)

dst = cv2.dilate(src=dst, kernel=None)

real_chess[dst>0.01 * dst.max()] = [255, 0, 0] # RGB
```

```python
plt.imshow(real_chess)
```

## Shi-Tomasi Corner Detection

```python
real_chess = cv2.imread('../Images/real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)

flat_chess = cv2.imread('../Images/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)

gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2GRAY)
gray_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_RGB2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(image=gray_flat_chess, maxCorners=64, qualityLevel=0.01, minDistance=10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x, y = i.ravel()
    cv2.circle(flat_chess, (x,y), 3, (255,0,0), -1)
```

```python
plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(image=gray_real_chess, maxCorners=100, qualityLevel=0.01, minDistance=10)
```

```python
corner = np.int0(corners)
```

```python
for i in corners:
    x, y = i.ravel()
    cv2.circle(real_chess, (x,y), 3, (255,0,0), -1)
```

```python
plt.imshow(real_chess)
```
