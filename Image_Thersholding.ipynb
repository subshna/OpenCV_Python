### Thersholding will convert an image to only two values, white or black

```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv2.imread('../Images/rainbow.jpg')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img, cmap='gray')
```

```python
img.max()
```

```python
ret, thersh1 = cv2.threshold(img, 127, 214, cv2.THRESH_BINARY) # 127 is half of color value 255, 255/2 = 127.5
```

```python
ret
```

```python
plt.imshow(thersh1, cmap='gray')
```

```python
ret, thersh2 = cv2.threshold(img, 127, 214, cv2.THRESH_BINARY_INV)
plt.imshow(thersh2, cmap='gray')
```

```python
ret, thersh2 = cv2.threshold(img, 127, 214, cv2.THRESH_TRUNC)
plt.imshow(thersh2, cmap='gray')
```

```python
ret, thersh3 = cv2.threshold(img, 127, 214, cv2.THRESH_TOZERO)
plt.imshow(thersh3, cmap='gray')
```

```python
img = cv2.imread('../Images/crossword.jpg', 0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15, 15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1 = cv2.threshold(img, 200, 255, cv2.THRESH_BINARY)
show_pic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 8)
show_pic(th2)
```

```python
blended = cv2.addWeighted(src1=th1, alpha=0.6, src2=th2, beta=0.4, gamma=0)
show_pic(blended)
```
