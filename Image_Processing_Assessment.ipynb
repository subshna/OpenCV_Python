## TASK: Open and display the giaraffes.jpg image that is located in the DATA folder.

```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_Img(img, cmap=None):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
img = cv2.imread('../Images/giraffes.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
display_Img(img)
```

## TASK:Apply a binary threshold onto the image.

```python
img.max()
```

```python
img = cv2.imread('../Images/giraffes.jpg', 0)
ret, thres = cv2.threshold(src=img, thresh=127, maxval=255, type=cv2.THRESH_BINARY)
```

```python
ret
```

```python
display_Img(thres, cmap='gray')
```

## TASK: Open the giaraffes.jpg file from the DATA folder and convert its colorspace to HSV and display the image.

```python
img = cv2.imread('../Images/giraffes.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
display_Img(img)
```

## TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)

```python
kernel = np.ones(shape=(4, 4), dtype = np.float32)/10
```

```python
kernel
```

```python
img = cv2.imread('../Images/giraffes.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
result = cv2.filter2D(src=img, ddepth=-1, kernel=kernel)
display_Img(result)
```

## TASK: Create a Horizontal Sobel Filter (sobelx from our lecture) with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.

```python
img = cv2.imread('../Images/giraffes.jpg', 0)
```

```python
sobelx = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)
```

```python
display_Img(sobelx, cmap='gray')
```

```python
img = cv2.imread('../Images/giraffes.jpg')

color = ['b', 'g', 'r']
for i, col in enumerate(color):
    histr = cv2.calcHist(images=[img], channels=[i], mask=None, histSize=[256], ranges=[0, 256])
    plt.plot(histr, color=col)
    plt.xlim([0, 256])
    
plt.title('HISTOGRAM OF GIRAFFES')
```
