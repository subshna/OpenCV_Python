```python
# Convert the image to HSL or HSV
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../Images/00-puppy.jpg')
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
plt.imshow(img)
```

```python
lst = ['cat', 'dog', 'tac', 'god', 'act']
```

```python
from itertools import combinations
```

```python
per_lst = combinations(lst, r=2)
for i, words in enumerate(per_lst):
    if (sorted(words[0]) == sorted(words[1])):
        print ('{} are anagrams'.format(words))
```
