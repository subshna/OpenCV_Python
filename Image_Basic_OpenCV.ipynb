```python
new_image = cv2.resize(fix_img, (1000, 400))
```

```python
plt.imshow(new_image)
```

```python
new_image.shape
```

```python
w_ratio = 0.8
h_ratio = 0.2 # 20% of original Height
```

```python
new_resize_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
```

```python
plt.imshow(new_resize_img)
```

```python
new_resize_img.shape
```

```python
new_flip_img = cv2.flip(fix_img, -1)
plt.imshow(new_flip_img) # 0 - Flip, 1 - flip with image reverse, -1 with both 0 and 1 flipped
```
